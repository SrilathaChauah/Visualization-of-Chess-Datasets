```python
#year_2017.groupby(['country'])['rating'].count().nlargest(5)
countrywise_players = year_2017.groupby(['country'])['rating'].count().nlargest(5).reset_index(name='Players')
countrywise_players
```

```python
countrywise_players_arrray = countrywise_players[["country"]].as_matrix().ravel()
countrywise_players_arrray
```

```python
countrywise_playerss_array = countrywise_players[["Players"]].as_matrix().ravel()
countrywise_playerss_array
```

```python
x = np.arange(countrywise_players.shape[0])
plt.bar(x, countrywise_playerss_array)
```

```python
x = np.arange(countrywise_players.shape[0])
plt.bar(x, countrywise_playerss_array)
plt.xticks(x, countrywise_players_arrray)
plt.ylabel("Number of players")
```
